## Pearson Correlation Coefficient  (R-value)


The pearson correlation coefficient (R-value) is used to identify "patterns" in variables. It is the most comman way to measure relationship in linear regression.

#### Positive pearson correlation coefficient
- The r value is between 0 and 1
- When one variable changes, the other variable will change in the same direction
- 0 < r < 0.3 is considered weak positive correlation 
- 0.3 < r < 0.5 is considered moderate positive correlation 
- 0.5 < r < 1 is considered strong positive correlation 

#### Negative pearson correlation coeffiecient
- The r value is between 0 and -1
- When one variable changes, the other variable wil change in the opposite direction
- 0 < r < -0.3 is considered weak negative correlation 
- -0.3 < r < -0.5 is considered moderate negative correlation 
- -0.5 < r < -1 is considered strong negative correlation 


#### No correlation
- The r value is 0
- There is no relationship between variables

#### The assumption made when using pearson correlation coefficient

 1. The variables are quantitative
Both dependant and independent variables are quantitative

----------------------------------
2. Normality\
The data follow normal distribution
----------------------------------
3. No outliers
There are no outliers inside both dependent and independent variable.
----------------------------------
4. Linear Relationship\
The relationship between the independent(s) variable and dependent variable are assumed to be linear.
----------------------------------

## Calculation of pearson correlation coefficient (r-value)

The calculation of pearson correlation coefficient, r is as below.

#### Import library

```python
# import library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# import library for linear regression

import statsmodels.api as sm

%matplotlib inline
```

#### Make dataset

```python
# create dataset and assign x and y
# we create an artificial data of bills(independent variable) and tips given(dependent variable)

x = [34, 108, 64, 88, 99 ,51] 
y = [5, 17, 11, 8, 14, 5]
df = pd.DataFrame({"bills" : x, "tips" : y})
x = df.bills
y = df.tips
df
```

```python
# scatterplot for x(independent variable) and y(dependent variable)
# see the trend using scatter plot
# we can notice upward trend between independent variable and dependent variable

plt.scatter(x, y, s  = 15 )
plt.xlabel("Bills amount ($)")
plt.ylabel("Tips amount ($)")
```

#### 1. Pearson correlation coefficient using python

```python
# count the required fields

x = [34, 108, 64, 88, 99 ,51] 
y = [5, 17, 11, 8, 14, 5]
df = pd.DataFrame({"x" : x, "y" : y})
df['x-x'] = df['x'] - np.mean(df['x'])
df['y-y'] = df['y'] - np.mean(df['y'])
df['(x-x)(y-y)'] = df['x-x'] * df['y-y']
df['(x-x)^2'] = df['x-x']**2
df['(y-y)^2'] = df['y-y']**2
df
```

```python
# pearson correlation coefficient

r_value = df['(x-x)(y-y)'].sum() / (df['(x-x)^2'].sum() * df['(y-y)^2'].sum()) ** 0.5
r_value
```

#### 2. Pearson correlation coefficient using statsmodels

```python
# add constant and use OLS fit model
# we have to add constant to X to fit into statsmodels.OLS

X = sm.add_constant(x)
model = sm.OLS(y,X).fit()
```

```python
#### Pearson correlation coefficient 

r_value = model.rsquared ** 0.5
print(r_value)
```
